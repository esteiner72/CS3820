```python
import pandas as pd
```

```python
df = pd.read_csv(r"data/Final_Augmented_dataset_Diseases_and_Symptoms.csv")
mappings = {}
index = 0
for disease in df["diseases"]:
    if disease not in mappings.keys():
        mappings[disease] = index
        index += 1
print(mappings.items())
```

```python
df.replace(mappings, inplace=True)
# Change on your laptop/machine if re-doing process (but we shouldn't need to)
df.to_csv(r"data\Final_Augmented_dataset_Diseases_and_Symptoms.csv", index=False)
```

```python
disease_counts = df.iloc[:, 0].value_counts()
print("Disease counts:\n", disease_counts)
```

Makes two lists, one for top 30 diseases and one for top 30 symptoms 

Trims the data keeping only the columns from the top 30 and only the symptoms from the top 30

```python

#THIS SECTION ISN'T WORKING QUITE RIGHT, IF YOU CAN FIGURE OUT HOW TO GET RID OF THE EXTRA COLUMN AT THE BEGINNING THAT WOULD BE GREAT
diseases_to_keep = disease_counts.head(30).index.tolist()
print(diseases_to_keep)

filtered_df = df[df['diseases'].isin(diseases_to_keep)]
print(filtered_df)

filtered_df.to_csv(r"data\Diseases_and_Symptoms_Filtered.csv", index=False)

#Finding the Most important symptoms
from sklearn.ensemble import ExtraTreesClassifier
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

data = pd.read_csv(r"data\Diseases_and_Symptoms_Filtered.csv")
X = data.iloc[:,1:379]  # Independent columns
y = data.iloc[:,0] 

bestfeatures = SelectKBest(score_func=chi2, k=30)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
scores = pd.concat([dfcolumns,dfscores],axis=1)
scores.columns = ['specs','score']

#append the sympts to a list and trimm all other sympts
top_features=scores.nlargest(30,'score')
symptoms_to_keep= ["diseases"]
symptoms_to_keep += top_features['specs'].tolist()

print(symptoms_to_keep)

trimmed_df = filtered_df[symptoms_to_keep]
trimmed_df = trimmed_df.reset_index(drop=True)  # This will reset any existing index
#trimmed_df = trimmed_df.drop(trimmed_df.columns[0], axis=1)  # Now drop the first column

trimmed_df.to_csv(r"data\Diseases_and_Symptoms_Filtered_Trimmed.csv", index=False)
```
